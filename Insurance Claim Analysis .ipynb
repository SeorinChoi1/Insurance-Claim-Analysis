```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/choeseolin/Downloads/Portofoilo/Data for Portofolio/insurance_data.csv")
df
```

```python
df=df.drop(columns = 'index')
df
```

```python
df=df.fillna('')
df
```

```python
df.sort_values(by="bmi", ascending=False).head(10)
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.rcParams['figure.figsize'] = (15,5)
plt.show()
```

```python
df=df.groupby('gender').mean(numeric_only=True).sort_values(by='claim', ascending= False)
df
```

```python
df.columns
```

```python
df2=df.transpose()
df2
```

```python
print(plt.style.available)
plt.style.use('ggplot')
```

```python
df2.plot(kind='bar', title=' Values By Gender Group')
plt.ylabel('Mean Value')
plt.xlabel('Features')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.transpose(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mean Feature Values by Gender (Heatmap)')
plt.xlabel('Gender')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```
